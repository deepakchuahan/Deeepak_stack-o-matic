Variable-Length Arguments

```python
#Write a solution that accepts a list of numbers and returns the median value

def find_median(numbers):
    # Sort the list first
    numbers.sort()

    n = len(numbers)

    if n % 2 == 1:
        return numbers[n // 2]
    else:
        middle1 = numbers[n // 2 - 1]
        middle2 = numbers[n // 2]
        return (middle1 + middle2) / 2

numbers = [12, 3, 5, 7, 19, 2]  #example
median = find_median(numbers)
print("The median is:", median)
```

```python
#Write a solution that accepts a variable number of arguments and returns the average of the numbers.

def calculate_average(*args):
    if len(args) == 0:
        return None

    return sum(args) / len(args)
average = calculate_average(10, 20, 30, 40, 50)
print("The average is:", average)
```

Error Handling

```python
#Handle the case where a division by zero occurs and print an appropriate error message

def calculate_average(*args):
    if len(args) == 0:
        print("Error: No numbers provided. Cannot calculate average.")
        return None

    return sum(args) / len(args)

average = calculate_average(10, 20, 30, 40, 50)
if average is not None:
    print("The average is:", average)

average = calculate_average()
```
